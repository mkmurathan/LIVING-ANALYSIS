```python
sns.barplot(data=df,x='Country',y='Cost_of_Living')
plt.xticks(rotation=20)
plt.title("Cost of Living Across Different Regions")
plt.show()
```

```python
melted_df = df.melt(id_vars='Country', value_vars=['Cost_of_Living', 'Average_Monthly_Income'], var_name='Category', value_name='Amount')

plt.figure(figsize=(14, 6))  # Grafik boyutunu ayarlama

# Barplot oluşturma
ax = sns.barplot(data=melted_df, x='Country', y='Amount', hue='Category')

# Her sütunun tepe noktasına değerleri yazdırma
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points')

plt.title("Cost of Living vs Average Monthly Income by Countries", fontdict={'fontsize': 18})
plt.legend(bbox_to_anchor=(1, 1))  # Legend pozisyonunu ayarlama
plt.show()
```

```python
custom_colors = ['#FF90BC', '#A9C46C', '#CDC1FF', '#B8001F']  # listing customized colors for the bars

# Veriyi gruplama ve ortalama hesaplama
basic_exp = df.groupby('Country')[['Housing_Cost_Percentage', 'Healthcare_Cost_Percentage', 'Transportation_Cost_Percentage', 'Education_Cost_Percentage']].mean().reset_index()

# Veriyi uzun formata dönüştürme
melted_basic_exp = basic_exp.melt(id_vars='Country', var_name='Expense Type', value_name='Expense Percentage')

# Grafik boyutunu ayarlama
plt.figure(figsize=(14, 6))

# Barplot oluşturma
ax = sns.barplot(data=melted_basic_exp, x='Country', y='Expense Percentage', hue='Expense Type', palette=custom_colors)

# Her sütunun tepe noktasına değerleri yazdırma
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'bottom', 
                xytext = (0, 5), 
                textcoords = 'offset points')

# Başlık ve etiketleri ayarlama
plt.title('Stacked Bar Plot for Basic Expenses by Countries', fontsize=18)
plt.xlabel('Country')
plt.ylabel('Expense Percentage')
plt.legend(title='Expense Type', bbox_to_anchor=(1, 1))

# Grafiği gösterme
plt.show()
```

```python

numerical_columns = [col for col in df.select_dtypes(include=['number']).columns if col != 'Year']   # selecting all numerical columns except year

grouped_df = df.groupby(['Year', 'Country'], as_index=False)[numerical_columns].mean()

sns.heatmap(grouped_df[numerical_columns].corr(),annot=True,fmt=".2f")

plt.title("Correlation Heatmap of Cost of Living Factors")
plt.show()
```

```python
Yearly_avg_cost= df.groupby('Year')['Cost_of_Living'].mean().reset_index()
print("Here are the average cost of living vlaues for each year:")
Yearly_avg_cost
```

```python
# Preparing line plot for overall cost of living trend:

plt.figure(figsize=(8,5))   # setting figure size

sns.lineplot(data=df, x='Year',y='Cost_of_Living',marker='o')  # plotting lineplot

max_point = Yearly_avg_cost.loc[Yearly_avg_cost['Cost_of_Living'].idxmax()]   # assigning the maximum and minimum values
min_point = Yearly_avg_cost.loc[Yearly_avg_cost['Cost_of_Living'].idxmin()]

plt.scatter(max_point['Year'] ,max_point['Cost_of_Living'], color='red', s=100, label='Maximum')   # plotting maximum and minimum points
plt.scatter(min_point['Year'] ,min_point['Cost_of_Living'], color='green', s=100, label='Maximum')

plt.text(max_point['Year'],max_point['Cost_of_Living'],f"Max: {max_point['Cost_of_Living']:.2f}")   # marking the values of maximum and minimum points
plt.text(min_point['Year'],min_point['Cost_of_Living'],f"Min: {min_point['Cost_of_Living']:.2f}")

plt.axvline(x=max_point['Year'], color='red', linestyle='--', linewidth=1)    # plotting vertical dotted lines at maximum and minimum
plt.axvline(x=min_point['Year'], color='green', linestyle='--', linewidth=1)

plt.text(max_point['Year'], 2000, f"{int(max_point['Year'])}", color='red',  rotation=90,ha='right'  )    # marking the year at maximum and minimum points
plt.text(min_point['Year'], 2000, f"{int(min_point['Year'])}", color='green',  rotation=90 ,ha='right' )

plt.title("Variation of Average Cost of Living with Year")
plt.show()
```

```python
fig,axes= plt.subplots(12,1,figsize=(12,20),sharex=True)    # creating subplots

countries= df['Country'].unique()

for i,country in enumerate(countries):                      # plotting for each country
    data =df[df['Country']==country]
    sns.lineplot(data=data,x='Year',y='Cost_of_Living',ax=axes[i])
    axes[i].set_title(country)

plt.tight_layout()     # adjust the space between subplots
plt.suptitle("Variation of Cost of Living for each Countries with Year",fontsize=18,y=1.01)   # common title for the subplots
plt.show()
```

```python
# Preparing data for scatter plot:

def income_cat(income):     # function to categorize income level
    if income <=2500:
        return 'low'
    elif income<=5000: 
        return 'medium'
    else:
        return 'high'
        
grouped_df['Income_Class']= grouped_df['Average_Monthly_Income'].apply(income_cat)   # column created for income class

size_mapping={'low': 50,'medium': 100,'high':200}
grouped_df['size']=grouped_df['Income_Class'].map(size_mapping)   # column for size of the point in scatter plot(based on income level)
```

```python
sns.scatterplot(data=grouped_df, x= 'Transportation_Cost_Percentage',y='Cost_of_Living',size='size')    # plotting scatter plot
sns.regplot(data=grouped_df, x='Transportation_Cost_Percentage',y='Cost_of_Living',scatter=False)       # plotting regression line

plt.legend(title='Income Class',handles=[                                                              # creating legend
    plt.Line2D([],[], marker='o',label='low',markersize=3,linestyle='None'),
    plt.Line2D([],[],marker='o',label='medium', markersize=5,linestyle='None'),
    plt.Line2D([],[], marker='o', label='high', markersize=8, linestyle='None')],
    bbox_to_anchor=(1.05, 1)) 

plt.title("Influence of Transportation Cost on  overall Cost of Living")
plt.show()
```

```python
sns.boxplot(data=grouped_df, x='Income_Class',y='Transportation_Cost_Percentage',order=['low','medium','high']) 

plt.title("Variation of Transportation Costs Across Income Classes")
plt.show()
```
